```python
#define a function  that takes list of number and return their sum
def sum_list(numbers):
    sum = 0
    for num in numbers:
        sum = sum + num
    return sum
sum_list([1, 2, 3, 4, 5])
```

```python
# write a function that finds largest number in a list
def find_largest(numbers):
    largest = numbers[0]   # assume first number is largest
    for num in numbers:
        if num > largest:
            largest = num
    return largest
find_largest([23,24,25,26,27])
```

```python
# create a funcion that takes a string  and return it in uppercase
def to_uppercase(text):
    return text.upper()
to_uppercase("hello world")
```

```python
#define  a function that counts  the number of vowels in a string
def count_vowels(text):
    vowels = "a, e, i, o, u, A, E, I, O, U"
    count = 0
    for char in text:
        if char in vowels:
            count = count+ 1
    return count
count_vowels("Hello World")
```

```python
#write a function that takes a number and return its factorial
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact = fact * i
    return fact
factorial(5)
```

```python
#write a function that cheak given  word is palindrome or not
def palindrome(word):
             
    if word == word[::-1]:      
        return "Palindrome"
    else:
        return "Not Palindrome"
palindrome("madam")
```

```python
#define a function that take two numbers and returns both their sum and difference
def sum_difference(a,b):
    sum = a+b
    difference = a-b
    return sum, difference
 
sum_difference(20,10)
    
```
